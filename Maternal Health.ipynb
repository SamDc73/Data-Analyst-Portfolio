## Dataset Overview

The Maternal_Indicators dataset contains information related to maternal health, including mortality rates, health indicators, and risk factors. The data is collected from various sites and spans multiple years, providing insights into trends and disparities in maternal health outcomes.

source of the dataset can be found on [Kaggle](https://www.kaggle.com/datasets/neharana404/maternal-indicators-in-us-states2016-2021)

## Prenatal Care and Behaviors:

### Early Prenatal Care Access

```python
%%sql

SELECT 
    Year,
    "Began prenatal care in 1st trimester - Weighted %" AS "Early Prenatal Care %"
FROM 
    Maternal_Indicators
GROUP BY 
    "Year"
ORDER BY 
    "Year";
```

- The percentage of women accessing early prenatal care shows minor fluctuations from year to year between 2016 and 2021, with a notable peak in 2020 

### Impact of Multivitamin Use on Maternal Mortality Rates

```python
%%sql

SELECT 
    CAST(REPLACE("Multivitamin use ≥4 times a week during the month before pregnancy - Weighted %", '%', '') AS INTEGER) AS "Multivitamin Use %",
    CAST(AVG(CAST("Deaths/100000" AS REAL)) + 0.999 AS INTEGER) AS "Average Maternal Mortality Rate"
FROM 
    Maternal_Indicators
GROUP BY 
    "Multivitamin Use %"
ORDER BY 
    "Multivitamin Use %" DESC;
```

- There seems to be a potential benefit in some cases at higher multivitamin usage rates, but this is not universally true across all the data.

- At the highest reported use rates (58% and 56%), the maternal mortality rates are relatively low (15 and 0 per 100,000, respectively), but there seems to be a potential benefit in some cases at higher multivitamin usage rates, but this is not universally true across all the data.


## Postpartum Care:

### State-by-State Postpartum Checkup Rates

```python
%%sql

SELECT 
    "Site Name",
    ROUND(AVG("Had maternal postpartum checkup - Weighted %"), 0) AS "Postpartum Checkup %"
FROM 
    Maternal_Indicators
GROUP BY
    "Site Name"
ORDER BY 
    "Postpartum Checkup %" DESC;

```

- High Engagement in Northern States: States like Rhode Island, New Hampshire, and Connecticut lead with the highest rates of postpartum checkups, each at 94%.

### Correlation Between Prenatal Care Initiation and Postpartum Depressive Symptoms

```python
%%sql

SELECT 
    ROUND("Began prenatal care in 1st trimester - Weighted %", 0) AS "Prenatal Care Initiation %",
    AVG(ROUND("Self-reported postpartum depressive symptoms‡ - Weighted %", 0)) AS "Average Postpartum Depressive Symptoms %"
FROM 
    Maternal_Indicators
GROUP BY 
    ROUND("Began prenatal care in 1st trimester - Weighted %", 0)
ORDER BY 
    "Prenatal Care Initiation %" DESC;

```

- Higher rates of prenatal care initiation (around 95% to 91%) do not consistently correspond with lower average rates of postpartum depressive symptoms, suggesting that early prenatal care alone may not significantly influence postpartum depression rates.


### Contraceptive use among postpartum women across different years

```python
%%sql

SELECT
    Year,
    ROUND("Use of any postpartum contraception§‖ - Weighted %", 0) AS "Postpartum Contraceptive Use %",
    ROUND("Highly effective contraceptive method: Male or female sterilization - Weighted %", 0) AS "Sterilization %",
    ROUND("Highly effective contraceptive method: Long-acting reversible contraceptive method¶ -Weighted %", 0) AS "LARC %",
    ROUND("Moderately effective contraceptive methods¶ - Weighted %", 0) AS "Moderately Effective Methods %",
    ROUND("Least effective contraceptive methods¶ - Weighted %", 0) AS "Least Effective Methods %"
FROM
    Maternal_Indicators
GROUP BY
    Year;
```

- The percentage of postpartum contraceptive use shows fluctuations over the years, with a notable decrease from 80% in 2016 to 77% in 2021.

- Use of long-acting reversible contraceptives (LARC) and sterilization varies, but the use of moderately effective methods has generally increased, peaking in 2019 and 2021 at 32%.


## Health Disparities:

### Variations in prenatal care access across geographic regions

```python
%%sql

SELECT
    "Site Name" AS Site,
    ROUND(AVG("Began prenatal care in 1st trimester - Weighted %"), 0) AS "Prenatal Care Initiation %"
FROM
    Maternal_Indicators
GROUP BY
    "Site Name"
ORDER BY
    "Prenatal Care Initiation %" DESC;
```

- Vermont and New Hampshire lead with the highest rates of prenatal care initiation at 93%, indicating strong early engagement in maternal healthcare in these states.

- The data shows a range from high initiation rates in the Northeast to lower rates in states like Texas at 80%

### Calculate the range and average of prenatal care initiation rates across all sites

```python
%%sql

SELECT
    ROUND(MAX("Began prenatal care in 1st trimester - Weighted %"), 0) AS "Max Prenatal Care Initiation %",
    ROUND(MIN("Began prenatal care in 1st trimester - Weighted %"), 0) AS "Min Prenatal Care Initiation %",
    ROUND(AVG("Began prenatal care in 1st trimester - Weighted %"), 0) AS "Average Prenatal Care Initiation %"
FROM
    Maternal_Indicators;
```

- The maximum prenatal care initiation rate across the dataset is 95%, while the minimum is significantly lower at 79%, indicating a wide disparity in early prenatal care engagement

### Geographic Distribution of Postpartum Depression Rates Across the U.S.

```python
%%sql

SELECT
    "Site Name" AS Site,
    ROUND(AVG("Self-reported postpartum depressive symptoms‡ - Weighted %"), 1) AS "Postpartum Depression %"
FROM
    Maternal_Indicators
GROUP BY
    "Site Name"
ORDER BY
    "Postpartum Depression %" DESC;
```

- The highest rates of postpartum depression are found in southern states, with Mississippi, Arkansas, and Alabama leading, showing rates of 22.0%, 20.4%, and 20.0% respectively.

- States like Iowa, Illinois, and Massachusetts report significantly lower rates of postpartum depression, with the lowest being Iowa at 9.4%.

## Insurance 

```python
%%sql
SELECT
    "Site Name" AS State,
    Year,
    ROUND("Private insurance - Weighted %", 0) AS "Private Insurance %",
    ROUND("Medicaid or CHIP/SCHIP - Weighted %", 0) AS "Medicaid or CHIP/SCHIP %",
    ROUND("No insurance - Weighted %", 0) AS "No Insurance %",
    CASE
        WHEN "Private insurance - Weighted %" > COALESCE("Medicaid or CHIP/SCHIP - Weighted %", 0) AND "Private insurance - Weighted %" > COALESCE("No insurance - Weighted %", 0) THEN 'Private Insurance'
        WHEN "Medicaid or CHIP/SCHIP - Weighted %" > COALESCE("Private insurance - Weighted %", 0) AND "Medicaid or CHIP/SCHIP - Weighted %" > COALESCE("No insurance - Weighted %", 0) THEN 'Medicaid or CHIP/SCHIP'
        WHEN "No insurance - Weighted %" > COALESCE("Private insurance - Weighted %", 0) AND "No insurance - Weighted %" > COALESCE("Medicaid or CHIP/SCHIP - Weighted %", 0) THEN 'No Insurance'
        ELSE 'Unknown'
    END AS "Most Used Insurance"
FROM
    Maternal_Indicators
    
ORDER BY
    State, Year;
```

## Smoking


### Trend of smoking before pregnancy over the years


```python
%%sql
SELECT
    Year,
    AVG(ROUND("Any cigarette smoking during the 3 months before pregnancy - Weighted %", 0)) AS "Smoking Before Pregnancy %"
FROM
    Maternal_Indicators
GROUP BY
    Year;
```

- There has been a consistent decrease in the percentage of women who reported smoking before pregnancy, from 19.5% in 2016 to 13.2% in 2021.

### Smoking rates before, during, and after pregnancy

```python
%%sql

SELECT
    Year,
    ROUND(AVG(CASE WHEN "Any cigarette smoking during the 3 months before pregnancy - Weighted %" IS NOT NULL THEN "Any cigarette smoking during the 3 months before pregnancy - Weighted %" ELSE 0 END), 0) AS "Smoking Before Pregnancy %",
    ROUND(AVG(CASE WHEN "Any cigarette smoking during the last 3 months of pregnancy - Weighted %" IS NOT NULL THEN "Any cigarette smoking during the last 3 months of pregnancy - Weighted %" ELSE 0 END), 0) AS "Smoking During Pregnancy %",
    ROUND(AVG(CASE WHEN "Any cigarette smoking postpartum - Weighted %" IS NOT NULL THEN "Any cigarette smoking postpartum - Weighted %" ELSE 0 END), 0) AS "Smoking Postpartum %"
FROM
    Maternal_Indicators

GROUP BY
    Year
```

- The reduction in smoking during pregnancy is notable

- On average, 52.56% of women who smoked before pregnancy stopped smoking during pregnancy.

- Of those who quit during pregnancy, approximately 35.46% resumed smoking postpartum.

### States with the highest and lowest smoking rates before pregnancy

```python
%%sql 

SELECT
    "Site Name" AS State,
    COALESCE(ROUND(AVG(CASE WHEN "Any cigarette smoking during the 3 months before pregnancy - Weighted %" IS NOT NULL THEN "Any cigarette smoking during the 3 months before pregnancy - Weighted %" ELSE 0 END), 0), 0) AS "Smoking Before Pregnancy %"
FROM
    Maternal_Indicators
GROUP BY
    "Site Name"
ORDER BY
    "Smoking Before Pregnancy %" DESC;
```
