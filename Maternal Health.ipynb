## Dataset Overview

This analysis utilizes a dataset containing maternal health indicators across various US states from the years 2016 to 2021. Below is a breakdown of the columns present in the dataset and their significance:

- **Site Name**: Represents the US state where the data was recorded.
- **Year**: Indicates the year in which the data was collected, ranging from 2016 to 2021.
- **Deaths/100000**: The maternal mortality rate per 100,000 live births in the state for the given year.
- **Multivitamin use ≥4 times a week during the month before pregnancy - Unweighted Sample Size**: The number of survey respondents.
- **Multivitamin use ≥4 times a week during the month before pregnancy - Unweighted Frequency**: The raw count of respondents who reported using multivitamins more than four times a week.
- **Multivitamin use ≥4 times a week during the month before pregnancy - Weighted %**: The percentage of respondents, adjusted for the population, who used multivitamins as described.
- **Multivitamin use ≥4 times a week during the month before pregnancy - Lower 95% Confidence Interval**: The lower bound of the 95% confidence interval for the percentage.
- **Multivitamin use ≥4 times a week during the month before pregnancy - Upper 95% Confidence Interval**: The upper bound of the 95% confidence interval for the percentage.
- **Underweight (body mass index [BMI] <18.5 kg/m2) - Unweighted Sample Size**: The number of respondents classified as underweight.
- **Underweight (body mass index [BMI] <18.5 kg/m2)**: The percentage of respondents classified as underweight.

This dataset provides a comprehensive view of maternal health trends and pre-pregnancy behaviors, including nutritional supplementation and body weight, across different states and over a series of years.
