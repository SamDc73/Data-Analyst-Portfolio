# Demographic Analysis

###  Age distribution of mothers

```python
%%sql

SELECT 
    CASE 
        WHEN MAGER9 = 1 THEN 'Under 15 years'
        WHEN MAGER9 = 2 THEN '15-19 years'
        WHEN MAGER9 = 3 THEN '20-24 years'
        WHEN MAGER9 = 4 THEN '25-29 years'
        WHEN MAGER9 = 5 THEN '30-34 years'
        WHEN MAGER9 = 6 THEN '35-39 years'
        WHEN MAGER9 = 7 THEN '40-44 years'
        WHEN MAGER9 = 8 THEN '45-49 years'
        WHEN MAGER9 = 9 THEN '50-54 years'
        ELSE 'Unknown'
    END AS age_group,
    COUNT(*) AS count,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS percentage
FROM 
    natality2022us
GROUP BY
    age_group

ORDER BY
    age_group
```

### Racial and ethnic distribution of mothers

```python
%%sql

SELECT 
    CASE 
        WHEN mrace6 = 1 THEN 'White'
        WHEN mrace6 = 2 THEN 'Black'
        WHEN mrace6 = 3 THEN 'American Indian/Alaska Native'
        WHEN mrace6 = 4 THEN 'Asian or Pacific Islander'
        ELSE 'Unknown or Not Stated'
    END AS race,
    COUNT(*) AS count,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS percentage
FROM 
    natality2022us
GROUP BY 
    race
ORDER BY 
    count DESC;
```

```python
%%sql

SELECT 
    CASE 
        WHEN mhisp_r = 0 THEN 'Non-Hispanic'
        WHEN mhisp_r = 1 THEN 'Hispanic'
        ELSE 'Unknown or Not Stated'
    END AS ethnicity,
    COUNT(*) AS count,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS percentage
FROM 
    natality2022us
GROUP BY 
    ethnicity
ORDER BY 
    count DESC;
```

### Education level of mothers

```python
%%sql

SELECT 
    CASE 
        WHEN meduc = 1 THEN '8th grade or less'
        WHEN meduc = 2 THEN '9th through 12th grade with no diploma'
        WHEN meduc = 3 THEN 'High school graduate or GED completed'
        WHEN meduc = 4 THEN 'Some college credit, but not a degree'
        WHEN meduc = 5 THEN 'Associate degree'
        WHEN meduc = 6 THEN 'Bachelor''s degree'
        WHEN meduc = 7 THEN 'Master''s degree'
        WHEN meduc = 8 THEN 'Doctorate or Professional degree'
        ELSE 'Unknown'
    END AS education_level,
    COUNT(*) AS count,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS percentage
FROM 
    natality2022us
GROUP BY 
    education_level
ORDER BY 
    count DESC;
```

### Marital status of mothers

```python
%%sql

SELECT 
    CASE 
        WHEN DMAR = 1 THEN 'Married'
        WHEN DMAR = 2 THEN 'Unmarried'
        ELSE 'Unknown or Not Stated'
    END AS marital_status,
    COUNT(*) AS count,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS percentage
FROM 
    natality2022us
GROUP BY 
    marital_status
ORDER BY 
    count DESC;
```

### Distribution of Births by Maternal Age Group 

```python
%%sql

SELECT 
  CASE 
    WHEN MAGER9 = 1 THEN 'Under 15 years'
    WHEN MAGER9 = 2 THEN '15-19 years'
    WHEN MAGER9 = 3 THEN '20-24 years'
    WHEN MAGER9 = 4 THEN '25-29 years'
    WHEN MAGER9 = 5 THEN '30-34 years'
    WHEN MAGER9 = 6 THEN '35-39 years'
    WHEN MAGER9 = 7 THEN '40-44 years'
    WHEN MAGER9 = 8 THEN '45-49 years'
    WHEN MAGER9 = 9 THEN '50-54 years'
    ELSE 'Unknown'
  END AS age_group,
  COUNT(*) AS birth_count
FROM natality2022us
GROUP BY age_group;
```

# Preterm Birth

### Percentage of Preterm Births (Less Than 37 Weeks Gestation)

```python
%%sql 

SELECT 
    SUM(CASE WHEN gestrec3 = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*) AS PretermBirthPercentage
FROM 
    natality2022us;
```

### Maternal Risk Factors Associated with Higher Rates of Preterm Birth

```python
%%sql

SELECT 
    rf_pdiab AS PrePregnancyDiabetes,
    rf_gdiab AS GestationalDiabetes,
    rf_phype AS PrePregnancyHypertension,
    rf_ghype AS GestationalHypertension,
    rf_ppterm AS PreviousPretermBirth,
    SUM(CASE WHEN gestrec3 = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*) AS PretermBirthRate
FROM 
    natality2022us
GROUP BY 
    rf_pdiab, rf_gdiab, rf_phype, rf_ghype, rf_ppterm;
```

### Preterm Birth Rates by Mother's Age

```python
%%sql

SELECT 
    mager AS MothersAge,
    SUM(CASE WHEN gestrec3 = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*) AS PretermBirthRate
FROM 
    natality2022us
GROUP BY 
    mager
ORDER BY 
    MothersAge;
```

### Preterm Birth Rates by Mother's Race/Ethnicity

```python
%%sql

SELECT 
    CASE 
        WHEN mrace6 = 1 THEN 'White (only)'
        WHEN mrace6 = 2 THEN 'AIAN (only)'
        WHEN mrace6 = 3 THEN 'American Indian/Alaska Native'
        WHEN mrace6 = 4 THEN 'Asian (only)'
        WHEN mrace6 = 5 THEN 'NHOPI (only)'
        WHEN mrace6 = 6 THEN 'More than one race'
    END AS MothersRaceEthnicity,
    SUM(CASE WHEN gestrec3 = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*) AS PretermBirthRate
FROM 
    natality2022us
GROUP BY 
    mrace6

```

### Preterm Birth Rates by Mother's Education Level

```python
%%sql

SELECT 
    CASE 
        WHEN meduc = 1 THEN '8th grade or less'
        WHEN meduc = 2 THEN '9th through 12th grade with no diploma'
        WHEN meduc = 3 THEN 'High school graduate or GED completed'
        WHEN meduc = 4 THEN 'Some college credit, but not a degree'
        WHEN meduc = 5 THEN 'Associate degree'
        WHEN meduc = 6 THEN 'Bachelor''s degree'
        WHEN meduc = 7 THEN 'Master''s degree'
        WHEN meduc = 8 THEN 'Doctorate or Professional degree'
        WHEN meduc = 9 THEN 'Unknown'
    END AS MothersEducation,
    SUM(CASE WHEN gestrec3 = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*) AS PretermBirthRate,
    AVG(CAST(mager AS INTEGER)) AS AvgMaternalAge,
FROM 
    natality2022us
GROUP BY 
    meduc

```

# Childbirth

### Proportion of Births in a Hospital vs. Out-of-Hospital

```python
%%sql

SELECT 
    CASE 
        WHEN bfacil = 1 THEN 'Hospital'
        WHEN bfacil IN (2, 3, 4, 5, 6, 7) THEN 'Out-of-Hospital'
        ELSE 'Unknown'
    END AS Birthplace,
    COUNT(*) AS BirthCount,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (), 2) AS BirthProportion
FROM 
    natality2022us
GROUP BY 
    Birthplace
ORDER BY 
    BirthProportion DESC;
```

### Rates of Cesarean Section Deliveries

```python
%%sql

SELECT 
    SUM(CASE WHEN dmeth_rec = 2 THEN 1 ELSE 0 END) AS CesareanCount,
    COUNT(*) AS TotalBirths,
    ROUND(SUM(CASE WHEN dmeth_rec = 2 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS CesareanRate
FROM 
    natality2022us
GROUP BY 
    dob_yy
ORDER BY 
    dob_yy;
```

### Differences in Delivery by Mother's Age

```python
%%sql

SELECT 
    CASE 
        WHEN mager9 = 1 THEN 'Under 15 years'
        WHEN mager9 = 2 THEN '15-19 years'
        WHEN mager9 = 3 THEN '20-24 years'
        WHEN mager9 = 4 THEN '25-29 years'
        WHEN mager9 = 5 THEN '30-34 years'
        WHEN mager9 = 6 THEN '35-39 years'
        WHEN mager9 = 7 THEN '40-44 years'
        WHEN mager9 = 8 THEN '45-49 years'
        WHEN mager9 = 9 THEN '50-54 years'
        ELSE 'Unknown'
    END AS MothersAgeGroup,
    SUM(CASE WHEN dmeth_rec = 1 THEN 1 ELSE 0 END) AS VaginalBirths,
    SUM(CASE WHEN dmeth_rec = 2 THEN 1 ELSE 0 END) AS CesareanBirths,
    COUNT(*) AS TotalBirths,
    ROUND(SUM(CASE WHEN dmeth_rec = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS VaginalRate,
    ROUND(SUM(CASE WHEN dmeth_rec = 2 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS CesareanRate
FROM 
    natality2022us
GROUP BY 
    mager9

```

### Differences in Delivery by Mother's Race/Ethnicity

```python
%%sql

SELECT 
    CASE 
        WHEN mrace6 = 1 THEN 'White (only)'
        WHEN mrace6 = 2 THEN 'AIAN (only)'
        WHEN mrace6 = 3 THEN 'American Indian/Alaska Native'
        WHEN mrace6 = 4 THEN 'Asian (only)'
        WHEN mrace6 = 5 THEN 'NHOPI (only)'
        WHEN mrace6 = 6 THEN 'More than one race'
    END AS MothersRaceEthnicity,
    SUM(CASE WHEN dmeth_rec = 1 THEN 1 ELSE 0 END) AS VaginalBirths,
    SUM(CASE WHEN dmeth_rec = 2 THEN 1 ELSE 0 END) AS CesareanBirths,
    COUNT(*) AS TotalBirths,
    ROUND(SUM(CASE WHEN dmeth_rec = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS VaginalRate,
    ROUND(SUM(CASE WHEN dmeth_rec = 2 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS CesareanRate
FROM 
    natality2022us
GROUP BY 
    mrace6
```

# Abnormal Conditions

### Most Common Abnormal Conditions:

```python
%%sql


SELECT 
    'Assisted ventilation required immediately after delivery' AS Condition,
    SUM(CASE WHEN ab_aven1 = 'Y' THEN 1 ELSE 0 END) AS Count,
    ROUND(SUM(CASE WHEN ab_aven1 = 'Y' THEN 1 ELSE 0 END) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS Percentage
FROM 
    natality2022us

UNION ALL

SELECT 
    'Assisted ventilation required for more than 6 hours' AS Condition,
    SUM(CASE WHEN ab_aven6 = 'Y' THEN 1 ELSE 0 END) AS Count,
    ROUND(SUM(CASE WHEN ab_aven6 = 'Y' THEN 1 ELSE 0 END) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS Percentage
FROM 
    natality2022us

UNION ALL

SELECT 
    'Admission to NICU' AS Condition,
    SUM(CASE WHEN ab_nicu = 'Y' THEN 1 ELSE 0 END) AS Count,
    ROUND(SUM(CASE WHEN ab_nicu = 'Y' THEN 1 ELSE 0 END) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS Percentage
FROM 
    natality2022us

UNION ALL

SELECT 
    'Surfactant replacement therapy' AS Condition,
    SUM(CASE WHEN ab_surf = 'Y' THEN 1 ELSE 0 END) AS Count,
    ROUND(SUM(CASE WHEN ab_surf = 'Y' THEN 1 ELSE 0 END) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS Percentage
FROM 
    natality2022us

UNION ALL

SELECT 
    'Antibiotics received by the newborn for suspected neonatal sepsis' AS Condition,
    SUM(CASE WHEN ab_anti = 'Y' THEN 1 ELSE 0 END) AS Count,
    ROUND(SUM(CASE WHEN ab_anti = 'Y' THEN 1 ELSE 0 END) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS Percentage
FROM 
    natality2022us

UNION ALL

SELECT 
    'Seizure or serious neurologic dysfunction' AS Condition,
    SUM(CASE WHEN ab_seiz = 'Y' THEN 1 ELSE 0 END) AS Count,
    ROUND(SUM(CASE WHEN ab_seiz = 'Y' THEN 1 ELSE 0 END) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS Percentage
FROM 
    natality2022us
    
UNION ALL 

SELECT 
    'No abnormal conditions' AS Condition,
    SUM(CASE WHEN NO_ABNORM = 1 THEN 1 ELSE 0 END) AS Count,
    ROUND(SUM(CASE WHEN NO_ABNORM = 1 THEN 1 ELSE 0 END) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS Percentage
FROM 
    natality2022us;
```

### Abnormal Conditions by Maternal Age Group

```python
%%sql

SELECT 
    CASE mager9
        WHEN 1 THEN 'Under 15 years'
        WHEN 2 THEN '15-19 years'
        WHEN 3 THEN '20-24 years'
        WHEN 4 THEN '25-29 years'
        WHEN 5 THEN '30-34 years'
        WHEN 6 THEN '35-39 years'
        WHEN 7 THEN '40-44 years'
        WHEN 8 THEN '45-49 years'
        WHEN 9 THEN '50-54 years'
        ELSE 'Unknown'
    END AS MothersAgeGroup,
    SUM(CASE WHEN ab_aven1 = 'Y' THEN 1 ELSE 0 END) AS AssistedVentilation,
    SUM(CASE WHEN ab_aven6 = 'Y' THEN 1 ELSE 0 END) AS AssistedVentilation6Hrs,
    SUM(CASE WHEN ab_nicu = 'Y' THEN 1 ELSE 0 END) AS NICUAdmission,
    SUM(CASE WHEN ab_surf = 'Y' THEN 1 ELSE 0 END) AS SurfactantTherapy,
    SUM(CASE WHEN ab_anti = 'Y' THEN 1 ELSE 0 END) AS AntibioticTherapy,
    SUM(CASE WHEN ab_seiz = 'Y' THEN 1 ELSE 0 END) AS Seizures
FROM 
    natality2022us
GROUP BY 
    mager9
```

### Infants Requiring NICU Admission and Assisted Ventilation by Maternal Age Group

```python
%%sql

SELECT 
    mager9 AS MothersAgeGroup,
    COUNT(*) AS TotalBirths,
    SUM(CASE WHEN ab_nicu = 'Y' THEN 1 ELSE 0 END) AS NICUAdmissions,
    ROUND(SUM(CASE WHEN ab_nicu = 'Y' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS NICUAdmissionRate,
    SUM(CASE WHEN ab_aven1 = 'Y' THEN 1 ELSE 0 END) AS AssistedVentilation,
    ROUND(SUM(CASE WHEN ab_aven1 = 'Y' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS AssistedVentilationRate
FROM 
    natality2022us
GROUP BY 
    mager9
ORDER BY 
    MothersAgeGroup;
```

# Prenatal Care

### distribution of when prenatal care begins among mothers

```python
%%sql

SELECT 
    CASE 
        WHEN PRECARE5 = 1 THEN '1st to 3rd month'
        WHEN PRECARE5 = 2 THEN '4th to 6th month'
        WHEN PRECARE5 = 3 THEN '7th to final month'
        WHEN PRECARE5 = 4 THEN 'No prenatal care'
        ELSE 'Unknown or not stated'
    END AS prenatal_care_start,
    COUNT(*) AS count,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS percentage
FROM 
    natality2022us
GROUP BY 
    prenatal_care_start
ORDER BY 
    count DESC;
```

### number of prenatal visits mothers had during their pregnancy.

```python
%%sql 


SELECT 
    CASE 
        WHEN PREVIS_REC = 1 THEN 'No visits'
        WHEN PREVIS_REC = 2 THEN '1 to 2 visits'
        WHEN PREVIS_REC = 3 THEN '3 to 4 visits'
        WHEN PREVIS_REC = 4 THEN '5 to 6 visits'
        WHEN PREVIS_REC = 5 THEN '7 to 8 visits'
        WHEN PREVIS_REC = 6 THEN '9 to 10 visits'
        WHEN PREVIS_REC = 7 THEN '11 to 12 visits'
        WHEN PREVIS_REC = 8 THEN '13 to 14 visits'
        WHEN PREVIS_REC = 9 THEN '15 to 16 visits'
        WHEN PREVIS_REC = 10 THEN '17 to 18 visits'
        WHEN PREVIS_REC = 11 THEN '19 or more visits'
        ELSE 'Unknown or not stated'
    END AS prenatal_visits,
    COUNT(*) AS count,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM natality2022us), 2) AS percentage
FROM 
    natality2022us
GROUP BY 
    prenatal_visits
ORDER BY
    percentage DESC;
```
